### Given above, distribution of # of initiated cases per member.

```python
# how many cases (by case_id and csm_pnd_dt) per member between 2016 and 2017.
sel = """
with temp as (
    select indiv_entpr_id, count(distinct case_id) as n_cid, count(distinct nvl(csm.csm_pnd_dt, csm.case_open_dt, csm.csm_accpt_dt )) as n_pnd_dt
    from ccw_csm_dtl as csm
    join analyst.INDIV_ENTERPRISE_ID_elig_2016to2017 as indiv on csm.indiv_entpr_id = indiv.indiv_enterprise_id
    where extract(year from to_date( nvl(csm.case_open_dt, csm.csm_pnd_dt,  csm.csm_accpt_dt ), 'DD-Mon-YY')) >=2016 
and extract(year from to_date( nvl(csm.case_open_dt, csm.csm_pnd_dt,  csm.csm_accpt_dt ), 'DD-Mon-YY')) <= 2017
    group by indiv_entpr_id
    )
select count(distinct indiv_entpr_id) as cnt_pid, n_cid, n_pnd_dt from temp
group by n_cid, n_pnd_dt
order by cnt_pid desc
;
"""

result = pd.read_sql(sel,con)

result.head(30)
```
