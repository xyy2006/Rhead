# Read in R model output varnames 

```python
topvars_pd = feather.read_dataframe("/gnshealthcare/shared/yyang/data/ses_data/topvars_char.feather")
```

```python
topvars_pd
```

## process varnames as some of them has extra postfix 

Like "_mergedLevels" and "_isNA_factor" were generated during feature QC/engineering stage. After removing them, calculate the uniq(list)

```python
topvars_pd.iloc[:,0].apply(lambda x : re.sub('(_\d+_mergedLevels|_isNA_factor)$', '', x))
```

```python
topvars_pd.iloc[:,0] = topvars_pd.iloc[:,0].apply(lambda x : re.sub('(_\d+_mergedLevels|_isNA_factor)$', '', x))
```

## split by cig_v or cig_c to two lists, since census and lifestyle tables are two tables in rawdata schema (wide format, query fast)

```python
census_topvars_lt = [re.sub('^cig_', '' , x) for x in topvars_pd.iloc[:,0] if re.search("cig_c", x)]
```

```python
lifestyle_topvars_lt = [re.sub('^cig_', '' , x)  for x in topvars_pd.iloc[:,0] if re.search("cig_v", x)]
```

```python
def count_duplicates(seq): 

    '''takes as argument a sequence and
    returns the number of duplicate elements'''

    return len(seq) - len(set(seq))

res = count_duplicates([-1,2,4,2,0,4,4])
print(res)  # -> 3
```

```python
count_duplicates(census_topvars_lt)
```

```python
count_duplicates(lifestyle_topvars_lt)
```

## SQL extract  census and lifestyle variables respectively

```python
#: test
#: census first, "c" initial words.
# join table to 100k members, instead of using python obj

sel = '''
select top 10 * from rawdata.census as c
right join analyst.origid_m31pop_100k_2 as m on c.indiv_enterprise_id = m.indiv_enterprise_id;
'''

census_tables = pd.read_sql(sel, con)
census_tables.head(10)

```

```python
#: cencus first, "c" initial words.

colname_select_part_char = ",".join( census_topvars_lt ) # for census
#origids_char = ",".join( m31pop_100k.origId.map(str) ) # already uploaded to database as a table much faster.
```

```python
# sel = '''
# select top 100 distinct indiv_enterprise_id, {0} from rawdata.census
# where indiv_enterprise_id in ({1});
# '''.format(colname_select_part_char, origids_char)

# census_tables = pd.read_sql(sel, con)
# census_tables.head(10)
```

```python
sel = '''
select distinct m.indiv_enterprise_id, {0} from rawdata.census as c
right join analyst.origid_m31pop_100k_2 as m on c.indiv_enterprise_id = m.indiv_enterprise_id
order by indiv_enterprise_id;
'''.format(colname_select_part_char)

census_tables = pd.read_sql(sel, con)
census_tables.head(10)
```

```python
census_tables.shape
```

```python
#: then lfstl_assmt, "v" initial words.

colname_select_part_char = ",".join( lifestyle_topvars_lt ) # for lfstl_assmt
#origids_char = ",".join( m31pop_100k.origId.map(str) )


sel = '''
select distinct m.indiv_enterprise_id, {0} from rawdata.lfstl_assmt as l
right join analyst.origid_m31pop_100k_2 as m on l.indiv_enterprise_id = m.indiv_enterprise_id
order by indiv_enterprise_id;
'''.format(colname_select_part_char)

lfstl_tables = pd.read_sql(sel, con)
lfstl_tables.head(10)

```

```python
lfstl_tables.shape
```

## Repeat  4.2 and 4.3 for Ning's model inferred variables and all 2016~2017 eligible members (as in "analyst.INDIV_ENTERPRISE_ID_elig_2016to2017")

```python
#: cencus first, "c" initial words.

colname_select_part_char = ",".join( census_topvars_lt ) # for census
#origids_char = ",".join( m31pop_100k.origId.map(str) ) # already uploaded to database as a table much faster.

# sel = '''
# select top 100 distinct indiv_enterprise_id, {0} from rawdata.census
# where indiv_enterprise_id in ({1});
# '''.format(colname_select_part_char, origids_char)

# census_tables = pd.read_sql(sel, con)
# census_tables.head(10)

sel = '''
select distinct elig.indiv_enterprise_id, {0} from rawdata.census as c
right join analyst.INDIV_ENTERPRISE_ID_elig_2016to2017 as elig on c.indiv_enterprise_id = elig.indiv_enterprise_id
order by indiv_enterprise_id;
'''.format(colname_select_part_char)

census_tables = pd.read_sql(sel, con)
census_tables.head(10)

census_tables.shape



#: then lfstl_assmt, "v" initial words.

colname_select_part_char = ",".join( lifestyle_topvars_lt ) # for lfstl_assmt
#origids_char = ",".join( m31pop_100k.origId.map(str) )


sel = '''
select distinct elig.indiv_enterprise_id, {0} from rawdata.lfstl_assmt as l
right join analyst.INDIV_ENTERPRISE_ID_elig_2016to2017 as elig on l.indiv_enterprise_id = elig.indiv_enterprise_id
order by indiv_enterprise_id;
'''.format(colname_select_part_char)

lfstl_tables = pd.read_sql(sel, con)
lfstl_tables.head(10)



lfstl_tables.shape
```
