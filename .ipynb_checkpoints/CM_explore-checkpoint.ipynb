#### Find intersections pairwise

```python
# Python program to illustrate the intersection
# of two lists
def intersection(lst1, lst2):
 
    # Use of hybrid method
    temp = set(lst2)
    lst3 = [value for value in lst1 if value in temp]
    return lst3
```

```python
# common_colnames_ccw_case_acty_n_case_dtl = np.intersect1d(colnames_ccw_case_acty, colnames_ccw_case_dtl)
# common_colnames_ccw_case_acty_n_case_dtl.tolist()
```

```python
table_name = "ccw_case_dtl"
```

```python
var = "SRC_CLSR_RSN_CD"
```

```python
ccw_dict['Column Physical Name'].apply(lambda x: x.lower())
ccw_dict['Column Physical Name'].map(lambda x: x.lower())
# both works
```

```python
for var in colnames_ccw_case_dtl:
    
    try: 
        anno = ccw_dict[ (ccw_dict.Table == table_name) & (ccw_dict['Column Physical Name'].map(lambda x: x.lower()) == var.lower())] \
        .iloc[0]['Column Definition'].strip()
    except LookupError:
        anno = "Not available in dict!"
#     else:
#         # for successful try
#         print("Good we found that!")

            
    
    sel = """
    select {0}, count( *) as cnt_row from rawdata.{1}
    group by {0}
    order by cnt_row desc
    limit 100
    ;
    """ .format(var, table_name)
    
    # cur.execute(sel)
    unique_value_content = pd.read_sql(sel,con)
    
    print("\n----------------------")
    print("Now {0}:".format(var))
    print("description: {0}".format(anno))    
    print(unique_value_content)
```

### Test query speed

```python
sel="""
SELECT INDIV_ENTERPRISE_ID, 
    sum(BILLD_CHRG_AMT) as BILLD_CHRG_AMT_SUM,
    sum(ELGBL_CHRG_AMT) as ELGBL_CHRG_AMT_SUM,
    sum(NET_PAYMT_AMT) as NET_PAYMT_AMT_SUM
FROM rawdata.clm_svc_fact 
GROUP BY INDIV_ENTERPRISE_ID
--LIMIT 50;
"""

t0 = time.time()

costs = pd.read_sql(sel,con)

costs.shape

costs.head()

costs.describe()

t1 = time.time()
print('Took %s seconds' %(t1-t0))
#much faster since it's already loaded in redshift, so no pd.read_csv 
```
